# **Sentiment Analysis on Resaerch Papers (Application)**

```python
from flask import Flask, render_template, request
from google.colab import files
import fitz
import pytesseract
import os
from flask import Flask, render_template, request, redirect, url_for
import uuid  # for generating unique folder names
from flask_ngrok import run_with_ngrok
import re
from textblob import TextBlob
app = Flask(__name__, template_folder='/content/drive/My Drive/LLM/')
run_with_ngrok(app)
lisst=[]
Dict={}
data=[]
pdf_path=""
text=""
output_folder=""
@app.route('/', methods=['GET', 'POST'])
def index():
    global text
    global pdf_path
    global output_folder
    message = None
    if request.method == 'POST':
        # check if the post request has the file part
        if 'file' not in request.files:
            message = 'No file part'
        else:
            file = request.files['file']
            # if user does not select file, browser also submit an empty part without filename
            if file.filename == '':
                message = 'No selected file'
            else:
                # save file to server
                UPLOAD_FOLDER = 'uploads'
                if not os.path.exists(UPLOAD_FOLDER):
                    os.makedirs(UPLOAD_FOLDER)
                file.save(os.path.join('uploads', file.filename))
                # open the PDF file
                UPLOAD_FOLDER = 'uploads'
                if not os.path.exists(UPLOAD_FOLDER):
                    os.makedirs(UPLOAD_FOLDER)
                pdf_path = os.path.join('uploads', file.filename)
                pdf = fitz.open(pdf_path)
                text=extract_text(pdf_path)
                # create a new folder for the images
                output_folder = os.path.join('/content/drive/My Drive/LLM/images', os.path.splitext(file.filename)[0])
                if not os.path.exists(output_folder):
                    os.makedirs(output_folder)
                    
                # iterate over pages and convert each page to image
                for i, page in enumerate(pdf):
                    pix = page.get_pixmap()
                    output_path = os.path.join(output_folder, f"image_{i}.png")
                    with open(output_path, "wb") as f:
                        f.write(pix.tobytes("png"))

                # close the PDF file
                pdf.close()

                message = 'File uploaded and converted successfully!'
                
    return render_template('main.html', message=message)

@app.route('/list')
def list():
    global lisst 
    global data
    global output_folder
    # Load the layout detection model
    colors = {'Text': 'blue', 'Title': 'green', 'List': 'red', 'Table': 'orange', 'Figure': 'yellow'}
    model = lp.Detectron2LayoutModel('lp://PubLayNet/mask_rcnn_X_101_32x8d_FPN_3x/config',
                                    extra_config=["MODEL.ROI_HEADS.SCORE_THRESH_TEST", 0.65],
                                    label_map={0: "Text", 1: "Title", 2: "List", 3:"Table", 4:"Figure"})

    # Initialize the OCR engine
    ocr_agent = lp.TesseractAgent(languages='eng')
    lis=[]
    # Load the NLP model
    nlp = spacy.load("en_core_web_trf")

    # Specify the folder containing the images
    image_folder = output_folder

    # Loop through all the image files in the folder
    for filename in os.listdir(image_folder):
        if filename.endswith(".jpg") or filename.endswith(".png"):
            # Load the image
            image_path = os.path.join(image_folder, filename)
            image = cv2.imread(image_path)
            image = image[..., ::-1] 
            layout = model.detect(image) # You need to load the image somewhere else, e.g., image = cv2.imread(...)
            lp.draw_box(image, layout,)
            text_blocks = lp.Layout([b for b in layout if b.type=="Title"])
            text_blocks
            figure_blocks = lp.Layout([b for b in layout if b.type=='Figure'])
            figure_blocks

            text_blocks = lp.Layout([b for b in text_blocks \
                              if not any(b.is_in(b_fig) for b_fig in figure_blocks)])
            h, w = image.shape[:2]

            left_interval = lp.Interval(0, w/2*1.05, axis='x').put_on_canvas(image)

            left_blocks = text_blocks.filter_by(left_interval, center=True)
            left_blocks.sort(key = lambda b:b.coordinates[1])

            right_blocks = [b for b in text_blocks if b not in left_blocks]
            right_blocks.sort(key = lambda b:b.coordinates[1])

            # And finally combine the two list and add the index
            # according to the order
            text_blocks = lp.Layout([b.set(id = idx) for idx, b in enumerate(left_blocks + right_blocks)])

            img = lp.draw_box(image, text_blocks, box_width=3, color_map=colors,show_element_id=False)
            img_array = np.asarray(img)  # Convert Image object to NumPy array
            cv2_imshow(img_array)
                # Initialize the tesseract ocr engine. You might need 
                # to install the OCR components in layoutparser:
            # pip install layoutparser[ocr]
            for block in text_blocks:
                segment_image = (block
                                  .pad(left=5, right=5, top=5, bottom=5)
                                  .crop_image(image))
                    # add padding in each image segment can help
                    # improve robustness 
                cv2_imshow(segment_image)
                text = ocr_agent.detect(segment_image)
                block.set(text=text, inplace=True)
            text_blocks.get_texts()
            lis.append(text_blocks.get_texts())
            sentence=["".join(txt) for txt in text_blocks.get_texts()]
            #print("Sentence = ",sentence)
            "".join(sentence)
            #lis.append("".join(sentence))
            #print("output = ",output)
            #lis.append(output)
    print(lis)
    lis = sum(lis, [])
    lis= [item.replace('\n\x0c', '') for item in lis]
    lis= [item.replace('\n', ' ') for item in lis]
    lis = [item for item in lis if item and item.strip()]
    cleaned_sentences = []
    for sentence in lis:
        cleaned_sentence = ''.join(c for c in sentence if not c in '!.').strip()
        cleaned_sentences.append(cleaned_sentence)
    data = [s.lstrip('0123456789') for s in cleaned_sentences]
    return render_template('list_check.html', data=data)

@app.route('/correct')
def correct():
    global data
    answer = request.args.get('answer')
    if answer == 'yes':
        message = 'Great, you got it right!'
    else:
        message = 'No Problem, We can fix it!!!'
        return redirect(url_for('new_list', list=request.args.getlist('data'), message=message))
    return message

@app.route('/new_list')
def new_list():
    global data
    message = request.args.get('message')
    data=data
    return render_template('correct.html', message=message, data=data)

@app.route('/newlist', methods=['POST'])
def newlist():
    global data
    correct_list = request.form['correct_list']
    data = correct_list.split('\n')
    return render_template('list_check.html', data=data)
@app.route("/senti", methods=["GET", "POST"])
def sentiment_A():
    global text
    global data
    global pdf_path
    global Dict
    text=extract_text(pdf_path)
    dataa= [item.replace('\r', '') for item in data]
    elements = dataa
    list_head = dataa
    list_head.append("Complete Research Paper")
    print(list_head)
    Dict = {}
    for i in range(0, len(list_head)):
        start_sentence = list_head[i]
        # Check if start_sentence is the last element of list_head
        if i == len(list_head) - 1 or start_sentence not in text:
            end_sentence = None
        else:
            end_sentence = list_head[i + 1]
        print("E = ", end_sentence)
        match = re.search(start_sentence, text)
        if match:
            start_index = match.end()
        else:
            start_index = 0
        # Check if end_sentence is None
        if end_sentence is None:
            result = text[start_index:]
        if end_sentence is not None:
            match = re.search(end_sentence, text)
            if match:
                end_index = match.start()
            else:
                end_index = len(text)
            result = text[start_index:end_index]
        key = start_sentence.strip()
        value = result.strip()
        Dict[key] = value
    if request.method == "POST":
        # Get the selected element
        selected_element = request.form.get("element")
        # Create a TextBlob object
        blob = TextBlob(Dict[selected_element])
        # Get the sentiment polarity (-1 to 1)
        polarity = blob.sentiment.polarity
        # Get the sentiment subjectivity (0 to 1)
        subjectivity = blob.sentiment.subjectivity
        # Render the template with the selected element, sentiment value, and index number
        # create a new SentimentIntensityAnalyzer object
        analyzer = SentimentIntensityAnalyzer()
        # analyze the text using the SentimentIntensityAnalyzer
        scores = analyzer.polarity_scores(Dict[selected_element])
        # print the scores for each sentiment category
        # print the proportion of positive, negative, and neutral words in the text
        positive=scores['pos']
        negative=scores['neg']
        neutral=scores['neu']
        polarity=round(polarity,4)
        subjectivity=round(subjectivity,4)    
        # Text dat
        # Create WordCloud object
        wordcloud = WordCloud(width=800, height=800, background_color='white', colormap=magma).generate(Dict[selected_element])

        # Create a buffer to store the image data
        img_buffer = io.BytesIO()

        # Save the WordCloud to the buffer
        plt.figure(figsize=(8,8))
        plt.imshow(wordcloud)
        plt.axis("off")
        plt.savefig(img_buffer, format='png')

        # Reset the buffer cursor to the beginning
        img_buffer.seek(0)

        # Convert image data to base64 for embedding in HTML
        img_str = base64.b64encode(img_buffer.getvalue()).decode('ascii')

        # Calculate word frequency
        word_count = Counter(Dict[selected_element].split())
        top_words = word_count.most_common(10)

        # Create bar chart of word frequency
        plt.figure(figsize=(8, 5))
        colors = plt.cm.plasma(np.linspace(0, 1,))  # Set colors using magma color map
        plt.bar([word[0] for word in top_words], [word[1] for word in top_words],color=colors)
        plt.xlabel('Word')
        plt.ylabel('Frequency')
        plt.title('Top 10 Words')
        plt.tight_layout()

        # Create a buffer to store the image data
        img_buffer = io.BytesIO()

        # Save the graph to the buffer
        plt.savefig(img_buffer, format='png')

        # Reset the buffer cursor to the beginning
        img_buffer.seek(0)

        # Convert image data to base64 for embedding in HTML
        graph_str = base64.b64encode(img_buffer.getvalue()).decode('ascii')
        nlp = spacy.load("en_core_web_sm")

        # Parse the text using spaCy
        doc = nlp(Dict[selected_element])

        # Use spaCy to generate a summary of the text
        summary = doc._.textrank_summary
        return render_template("sentiment.html", elements=elements, wordcloud=img_str, graph=graph_str, summary=summary,selected_element=selected_element, Dict=Dict, sentiment=polarity, subjectivity=subjectivity,pos=positive,neg=negative,neu=neutral)
    else:
          # Render the template with the list of elements
        return render_template("sentiment.html", elements=elements)



if __name__ == '__main__':
    app.run()
    
```
